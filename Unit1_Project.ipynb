Motivational Quote Generator

Goal: Generate unique motivational quotes based on keywords like "Success", "Grind", "Peace".

Tech: gpt2 with high temperature for creativity.

Tech used: pipeline('text-generation')


1. Setup and Model Initialization

Here, we load the gpt2 model into a text-generation pipeline. This cell handles the "heavy lifting" of downloading the model weights.


```python
from transformers import pipeline, set_seed

# Initialize the pipeline
# We use 'gpt2' which is lightweight and perfect for creative text generation
quote_generator = pipeline('text-generation', model='gpt2')

# Setting a seed ensures your results are reproducible for your submission
set_seed(42)

print("Model loaded successfully!")
```

2. The Quote Generation Logic

This cell contains the core function. I’ve added parameters like top_k and repetition_penalty to ensure the quotes are high-quality and don't repeat themselves.

```python
def generate_motivational_quote(keyword):
    # Creating a structured prompt to guide GPT-2
    prompt = f"A wise person once said about {keyword}: \""
    
    # Generation parameters:
    # temperature: Higher (0.9) for more creativity
    # max_new_tokens: Limits the length to keep it 'quote-sized'
    # repetition_penalty: Prevents the model from getting stuck in loops
    sequences = quote_generator(
        prompt,
        max_new_tokens=40,
        temperature=0.9,
        top_k=50,
        top_p=0.95,
        repetition_penalty=1.2,
        do_sample=True,
        num_return_sequences=1,
        pad_token_id=50256 # GPT-2 specific end-of-sentence token
    )
    
    # Post-processing to clean up the output
    raw_text = sequences[0]['generated_text']
    # We strip the prompt to show only the quote, and try to stop at the first closing quote
    quote_only = raw_text.split('"')[1] if '"' in raw_text.split(f"about {keyword}:")[1] else raw_text
    
    return quote_only.strip()

print("Function 'generate_motivational_quote' is ready.")
```

3. Testing the Generator

This is the cell you can use to demonstrate your project during the submission.

```python
# Define your keywords
keywords = ["Success", "Grind", "Peace"]

print("--- Creative Motivational Quotes ---\n")
for kw in keywords:
    quote = generate_motivational_quote(kw)
    print(f"Keyword: {kw}")
    print(f"Quote: \"{quote}\"")
    print("-" * 30)
```

4. User Input Interface

```python
# Simple interactive loop
user_kw = input("Enter a keyword for your quote (e.g., Courage, Hustle, Calm): ")
if user_kw:
    print(f"\nGenerated Quote: \"{generate_motivational_quote(user_kw)}\"")
```
