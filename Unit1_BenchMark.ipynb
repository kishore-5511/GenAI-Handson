```python
import nltk
from transformers import pipeline, set_seed

nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')

set_seed(42) # Keeps results consistent
```

Experiment 1: Text Generation
The Goal: See what happens when you ask an "understanding" model (BERT) to "write" text.

```python
models = ['bert-base-uncased', 'roberta-base', 'facebook/bart-base']
prompt = "The future of Artificial Intelligence is"

for model_name in models:
    print(f"\n--- Testing Generation with: {model_name} ---")
    try:
        # Note: BERT/RoBERTa may throw a warning or error here 
        # because they aren't 'Text Generation' models.
        generator = pipeline('text-generation', model=model_name)
        result = generator(prompt, max_length=20, num_return_sequences=1)
        print(result[0]['generated_text'])
    except Exception as e:
        print(f"FAILED: {model_name} cannot generate text easily. Error: {e}")
```

Experiment 2: Masked Language Modeling (MLM)
The Goal: This is the "home turf" for BERT and RoBERTa.

```python
# Define the prompts for each specific model type
prompt_bert = "The goal of Generative AI is to [MASK] new content."
prompt_roberta_bart = "The goal of Generative AI is to <mask> new content."

# 1. BERT Test (Uses [MASK])
fill_bert = pipeline('fill-mask', model='bert-base-uncased')
print("BERT:", fill_bert(prompt_bert)[0]['token_str'])

# 2. RoBERTa Test (Uses <mask>)
fill_roberta = pipeline('fill-mask', model='roberta-base')
print("RoBERTa:", fill_roberta(prompt_roberta_bart)[0]['token_str'])

# 3. BART Test (Uses <mask>)
fill_bart = pipeline('fill-mask', model='facebook/bart-base')
print("BART:", fill_bart(prompt_roberta_bart)[0]['token_str'])
```

Experiment 3: Question Answering
The Goal: Observe the difference between a "Base" model and one that has been "Fine-tuned."

```python
context = "Generative AI poses significant risks such as hallucinations, bias, and deepfakes."
question = "What are the risks?"

for model_name in models:
    print(f"\n--- Testing QA with: {model_name} ---")
    try:
        qa_pipe = pipeline('question-answering', model=model_name)
        res = qa_pipe(question=question, context=context)
        print(f"Answer: {res['answer']}")
    except Exception as e:
        print(f"Model {model_name} struggled. Error: {e}")
```
